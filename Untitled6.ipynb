Atividade 1 -  classificação de número

```python
#Atividade 01
numero = float(input('digite um número:'))
if numero > 0:
  print('numero positivo')
elif numero == 0:
  print('numero nulo')
else:
  print('numero negativo')
```

```python
#Atividade 02
numero = int(input('digite um número:'))
if numero % 2 == 0:
  print('0 número é par')
else:
  print('0 número é impar:')
```

```python
#Atividade 03
nota = float(input('digite sua nota:'))
if nota >=7:
  print('Aprovado')
elif nota >=5:
  print('Recuperação')
else:
  print('Reprovado')
```
